```python
dic=nltk.corpus.words.words ('en')
print(type(dic))
```

dic(타입은 list)에서 for loop를 돌려서 하나씩 w로 보낸다. 
그때 if ~ 조건이 적용되어 돌아간다. 즉, w 에 들어간 것이 소문자이면, 맨 앞의 w로 보낸다
그것들이 wordlist에 쌓인다.

```python
wordlist = [w for w in dic if w.islower()]
print(wordlist[:10])
```

그걸 내용을 보여줄수도 있고 인덱스 즉, 몇 번에 있는지 보여줄수도 있다

즉, 여기에서는 -ed로 끝나는 단어만 result에 쌓인다.

처음부터 10개까지만 보여지도록 설정했기 때문에, print (result) 와 print (result[:10])의 결과값은 같다

정규 표현안에 []이 있으면 [ghi]의 경우g,h,i중 뭐가 나와도 상관없다는 의미이다
네 개의 캐릭터만 허용된다. dddd 이렇게 나오는 건 안되고 하나의 []에서 하나씩 골라서 총 네 개의 캐릭터가 온다

```python
result = [w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)][:10]
print(result[:10])
```

^[ah]+$에서 +의 의미는? one or more... 반복해서 나올 수 있다는 뜻이다. 하나가 나와도 된다

즉, a 혹은 h 인데 반복되서 나와도 된다는 뜻이다

```python
result = [w for w in wordlist if re.search('^[ah]+$', w)][:10]
print(result[:10])
```

wsj 는 word street journal 에 나왔던 모든 단어들이다. 여기서도 variable type은 list 다.
따라서 for loop를 통해서 하나하나 적용시켜줘야 한다. 
각 개체는 하나하나씩 str 이지만 전체는 list 이므로 각 str에 한번씩 적용되도록 for loop를 설정한다.

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
wsj
```

[0-9]은 모든 숫자이다 0에서 9를 의미한다. 
+를 붙인 것은 하나이상 숫자가 나와도 된다는 뜻을 주기 위함이다.
그 다음에 나오는 .은 그냥.을 쓰면 정규표현에서의 캐릭터 하나가 된다
때문에.을 찾아라 라고 설정하고 싶을 때는 특수기호를 이용해야 한다. 
즉, 정규표현에서의 . 과 중복되지 않게 하기 위함이다. 
그럴때는 \하고 .을 써준다. 예를 들어 +를 찾고 싶다고 할때도 \와 +를 같이 써주면 된다. 
 
'^[0-9]+\.[0-9]+$'결국 이것은 실수를 말한다. 숫자다. 숫자가 무엇이든 나와도 된다는 표현이다

```python
wordlist = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
print(wordlist[:10])
```

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```
