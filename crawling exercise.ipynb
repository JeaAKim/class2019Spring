```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen('https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=105&date=20190322')
bsObject = BeautifulSoup(html, "html.parser")
```

```python
bsObject
```

```python
for n in range (0,25):
    li = bsObject.find("li", {"class":"ranking_item is_num"+str(n+1)})
    title = li.select('a')[0].get('title')
    press = li.next_sibling.next_sibling.select('div')[-2].get_text()
    view = li.next_sibling.next_sibling.select('div')[-1].get_text()
    print(n+1,title,press,view)
```

```python
with open('crawling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'press', 'view'])
        
    for n in range(0,25):
        li = bsObject.find("li", {"class":"ranking_item is_num"+str(n+1)})
        title = li.select('a')[0].get('title')
        press = li.next_sibling.next_sibling.select('div')[-2].get_text()
        view = li.next_sibling.next_sibling.select('div')[-1].get_text()    
        writer.writerow([n+1, title,press, view])
```
